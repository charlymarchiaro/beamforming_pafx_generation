```python
ang_deg = range(0, 360, 1)
ang = [a * math.pi / 180 for a in ang_deg]
r = [0.5 * math.sin(9 * a) + 1 for a in ang]
x = []
y = []
for i, a_i in enumerate(ang):
    r_i = r[i]
    x_i = r_i * math.cos(a_i)
    y_i = r_i * math.sin(a_i)
    x.append(x_i)
    y.append(y_i)
plt.figure(figsize=(8, 8))
plt.plot(x, y)
```
